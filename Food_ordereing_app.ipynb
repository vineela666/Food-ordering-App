```python
class FoodOrderingApp:
    def __init__(self):
        self.users = {}
        self.admins = {}

    def register_user(self, username, password):
        if username in self.users:
            print("Username already exists! Please choose another one.")
            return False
        else:
            self.users[username] = password
            print("User registered successfully!")
            return True

    def register_admin(self, username, password):
        if username in self.admins:
            print("Admin username already exists! Please choose another one.")
            return False
        self.admins[username] = password
        print("Admin registered successfully!")
        return True

    def user_login(self, username, password):
        if username in self.users and self.users[username] == password:
            print("User logged in successfully!")
            return True
        else:
            print("Invalid username or password!")
            return False

    def admin_login(self, username, password):
        if username in self.admins and self.admins[username] == password:
            print("Admin logged in successfully!")
            return True
        else:
            print("Invalid admin username or password!")
            return False

class Restaurant:
    def __init__(self, name, menu):
        self.name = name
        self.menu = menu

    def display_menu(self):
        for item, price in self.menu.items():
            print(f"{item}: ${price:.2f}")

    def place_order(self):
        order = {}
        while True:
            item = input("Enter the item you want to order (or 'done' to finish): ").capitalize()
            if item == "Done":
                break
            if item in self.menu:
                quantity = int(input(f"Enter the quantity of {item}: "))
                order[item] = quantity
            else:
                print("Sorry, that item is not on the menu.")
        return order

    def calculate_bill(self, order):
        total_bill = 0
        for item, quantity in order.items():
            if item in self.menu:
                total_bill += self.menu[item] * quantity
        return total_bill

def main():
    app = FoodOrderingApp()

    while True:
        print("\nWelcome to Food Ordering App")
        print("1. Register as User")
        print("2. Register as Admin")
        print("3. User Login")
        print("4. Admin Login")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            username = input("Enter username: ")
            password = input("Enter password: ")
            app.register_user(username, password)
        elif choice == "2":
            username = input("Enter admin username: ")
            password = input("Enter admin password: ")
            app.register_admin(username, password)
        elif choice == "3":
            username = input("Enter your username: ")
            password = input("Enter your password: ")
            app.user_login(username, password)
        elif choice == "4":
            username = input("Enter admin username: ")
            password = input("Enter admin password: ")
            app.admin_login(username, password)
        elif choice == "5":
            print("Exiting...")
            break
        else:
            print("Invalid choice! Please choose again.")

    menu = {"Burger": 5.99, "Pizza": 8.99, "Pasta": 7.49, "Salad": 4.99, "Drink": 1.99}

    restaurant = Restaurant("Example Restaurant", menu)

    print("Welcome to", restaurant.name)
    print("Here is our menu:")
    restaurant.display_menu()

    order = restaurant.place_order()
    print("\nYour order:")
    for item, quantity in order.items():
        print(f"{item}: {quantity}")

    total_bill = restaurant.calculate_bill(order)
    print(f"\nTotal Bill: ${total_bill:.2f}")

if __name__ == "__main__":
    main()
```
